1. Uma tabela onde tenha nas colunas os individuos, e nas linhas os CIGAR_ID, e as celulas são preenchidas com 1 ou 0, dependendo se aquele individuo possui o CIGAR_ID ou não.


```python
def split_cigar_ids(data):
    """
  Divide os CIGAR_IDs por vírgula e retorna uma lista de listas.

  Args:
    data: O dataframe de dados.

  Returns:
    Uma lista de listas, onde cada lista interna representa os CIGAR_IDs para uma determinada linha do dataframe.
    """
    cigar_ids_list = []
    n = len(data.columns)

    for row in df.iterrows():
        #print(row[1][0])

        for i in range(1,n-2):
            #print(row[1][i])
            #print("##########\n")
            if row[1][i] != 0:
                cigar_ids = row[1][i].split(",")

                for val in cigar_ids:
                    if val != '':
                        index = val.find("(")
                        val = val[:index]
                        cigar_ids_list.append(val)

    return np.unique(cigar_ids_list)
```

```python
all_cigars = split_cigar_ids(df)
all_cigars = np.char.replace(all_cigars, " ", "") #tira os espaços dos nomes

final_all_cigars = [x for x in all_cigars if x != ""] #elimina posições vazias
final_all_cigars = np.unique(final_all_cigars)
print(final_all_cigars)
```

```python
presence_absence_table = pd.DataFrame(columns=df.columns,index=final_all_cigars)
#presence_absence_table.index = final_all_cigars
presence_absence_table = presence_absence_table.drop(['locus','num_padroes', 'qntd_informa'], axis=1)
presence_absence_table
```

```python
cols = presence_absence_table.columns
rows = presence_absence_table.index

for row in rows: #percorre as linhas

    for col in cols: # percorre as colunas (mesmo indice do df)
        k = 0
        for values in df[col]: # verifica se o cigar_id esta na coluna da cultivar
            #print(row,values)
            if values != 0:
                if row in values:
                    k = 1

        if k == 1: # se esta, adiciono 1 ao novo dataset
            presence_absence_table.loc[row,col] = 1
```

```python
presence_absence_table = presence_absence_table.fillna(0)  # Replace all NaN values with 0
print(presence_absence_table)
```

# Parte 2

2. Uma tabela onde tenha nas colunas os individuos, e nas linhas os CIGAR_ID, e as celulas são preenchidas com a cobertura.


```python
# Criar uma tabela da cobertura dos CIGAR_IDs. Mesma estrutura da anterior.
coverage_table = pd.DataFrame(columns=df.columns,index=final_all_cigars)
coverage_table = coverage_table.drop(['locus','num_padroes', 'qntd_informa'], axis=1)
coverage_table
```

```python
# Criar uma tabela da cobertura dos CIGAR_IDs. Mesma estrutura da anterior.
scaffold_table = pd.DataFrame(columns=df.columns,index=final_all_cigars)
scaffold_table = scaffold_table.drop(['locus','num_padroes', 'qntd_informa'], axis=1)
scaffold_table
```

```python
import re
```

```python
cols = coverage_table.columns
rows = coverage_table.index

for row in rows: #percorre as linhas

    for col in cols: # percorre as colunas (mesmo indice do df)
        k = 0
        coverages = []
        scaff = []
        #print("########:",col)
        for index,values in enumerate(df[col]): # verifica se o cigar_id esta na coluna da cultivar
            #print(row,values)
            if values != 0 and row in values:
                    #print(row,values,df.loc[index,'locus'])
                    scaff.append(df.loc[index,'locus'])
                    regex = rf"({row})(.*?)(?=\))"
                    #print(regex)
                    match = re.search(regex, values)

                    if match:
                      value = match.group(2).lstrip("(")
                      #print(value)
                      coverages.append(value)

                    k=1

        if k == 1: # se esta, adiciono as infos
            scaffold_table.loc[row,col] = scaff
            coverage_table.loc[row,col] = coverages
    #break
```

A tabela de coverage armazena informações extras de cobertura.

3. Uma tabela onde tenha nas colunas os scaffolds, e nas linhas os CIGAR_ID, e as celulas são preenchidas com 1 ou 0, dependendo se aquele scaffold possui o CIGAR_ID ou não.

```python
# Criar uma tabela de resença do cigar_id. Mesma estrutura da anterior.
scaffold_presence_absence_table = pd.DataFrame(columns=final_all_cigars,index=df['locus'])
scaffold_presence_absence_table.head(3)
```

```python
# Criar uma tabela de fenotipos por scaffold e cigar_id. Mesma estrutura da anterior.
fenotipo_presence_absence_table = pd.DataFrame(columns=final_all_cigars,index=df['locus'])
fenotipo_presence_absence_table.head(3)
```

```python
cols = scaffold_presence_absence_table.columns
rows = scaffold_presence_absence_table.index

for row in rows: #percorre as linhas

    for col in cols: # percorre as colunas (mesmo indice do df)
        k = 0
        phenotypes = []

        for index,values in enumerate(df['locus']): # verifica se o cigar_id esta na linha do scaffold
            #print("###",row,values,col)
            if row == values:
              #print(row,values,col)
              valores_row = df[df["locus"] == row].squeeze()
              valores_row_str = [str(valor) for valor in valores_row.values]  # transforma em str pro join funcionar
              valores_row_str = '\t'.join(valores_row_str)
              #print(col,valores_row)
              if col in valores_row_str: #verifica se a substring esta na string
                k=1

              for coluna in scaffold_table.columns:
                #print("#######",col,coluna)
                #print(valores_row)
                if valores_row[coluna]!=0:
                  if col in valores_row[coluna]:
                    #print(row,col,coluna)
                    phenotypes.append(coluna)

        if k == 1: # se esta, adiciono as infos
            scaffold_presence_absence_table.loc[row,col] = 1
            fenotipo_presence_absence_table.loc[row,col] = phenotypes
```
